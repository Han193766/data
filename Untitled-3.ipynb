```python
categorical_features = ['condition','yr_built','yr_renovated']
numerical_features = ['price','sqft_above','sqft_basement']
target = 'id'
```

```python
# split data in train / test
X_train, X_test, y_train, y_test = train_test_split(df[categorical_features + numerical_features]
                                                    ,df[target],
                                                    test_size = 0.3, random_state=1234)
```

```python
# define transformers
cat_transformer = Pipeline(steps = [('onehot', OneHotEncoder(handle_unknown='ignore'))])
num_transformer = Pipeline(steps = [('scaler', StandardScaler())])
```

```python
# building processor
preprocessor = ColumnTransformer(transformers = [('cat',cat_transformer, categorical_features),
                                                 ('num', num_transformer, numerical_features)])
```

```python
linear_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
linear_regression.fit(X_train, y_train)
```

```python
pred = linear_regression.predict(X_test)
```

```python
pred
```

```python
np.sqrt(mean_squared_error(pred,y_test))
```

```python
ridge_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', Ridge(alpha=3.0))
])
```

```python
ridge_regression.fit(X_train, y_train)
```

```python
pred_ridge = ridge_regression.predict(X_test)
```

```python
np.sqrt(mean_squared_error(pred_ridge,y_test))
```

```python
lasso_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', Lasso(alpha=1.0))
])
```

```python
lasso_regression.fit(X_train, y_train)
```

```python
lasso_pred = lasso_regression.predict(X_test)
```

```python
np.sqrt(mean_squared_error(lasso_pred, y_test))
```

```python
elasticnet_regression = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('regressor', ElasticNet(alpha=1.0, l1_ratio=0.2))
])
```

```python
elasticnet_regression.fit(X_train, y_train)
```

```python
elasticnet_pred = elasticnet_regression.predict(X_test)
```

```python
np.sqrt(mean_squared_error(elasticnet_pred, y_test))
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df
```

```python
df['log_price'] = np.log1p(df['price'])

# bathroom
df['total_bathrooms'] = df['bedrooms'] + df['bathrooms']

# sqft_living
df['log_sqft_living'] = np.log1p(df['sqft_living'])
df['log_sqft_lot'] = np.log1p(df['sqft_lot'])

# floor
df['multi_floor'] = np.where(df['floors'] > 1, 1, 0)

# waterfront$view
df['waterfront'] = df['waterfront'].astype('category')
df['view'] = df['view'].astype('category')

# year

df['renovated'] = np.where(df['yr_renovated'] > 0, 1, 0)

# feagures
selected_features = ['bedrooms', 'bathrooms', 'total_bathrooms', 'log_sqft_living', 'log_sqft_lot',
                     'grade', 'waterfront', 'view', 'multi_floor', 'renovated']

# std
numerical_features = ['bedrooms', 'bathrooms', 'total_bathrooms', 'log_sqft_living', 'log_sqft_lot', 'grade']
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# print
print(df[selected_features].head())
```
