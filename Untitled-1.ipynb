# start to calculate p-value and t-test

```python
import numpy as np
from scipy import stats
```

```python
from scipy import stats

# Split the data into two groups
checkout_yes = df[df['checkout_status'] == 1]['time_spent']
checkout_no = df[df['checkout_status'] == 0]['time_spent']

# Perform the t-test
t_stat, p_value = stats.ttest_ind(checkout_yes, checkout_no)

p_value
```

```python
# Display the t-statistic
t_stat
```

```python
# List of numerical features
num_features = ['age', 'sessions', 'pages_visited', 'cart_items', 'cart_value']

# Perform t-tests for other numerical variables
t_stats = []
p_values = []

for feature in num_features:
    checkout_yes = df[df['checkout_status'] == 1][feature]
    checkout_no = df[df['checkout_status'] == 0][feature]
    t_stat, p_value = stats.ttest_ind(checkout_yes, checkout_no)
    t_stats.append(t_stat)
    p_values.append(p_value)

# Create a DataFrame to display the results
t_test_results = pd.DataFrame({
    'Feature': num_features,
    't-statistic': t_stats,
    'p-value': p_values
})

# Perform chi-square tests of independence for categorical variables
from scipy.stats import chi2_contingency

cat_features = ['device', 'location']
chi2_stats = []
p_values_chi2 = []

for feature in cat_features:
    contingency_table = pd.crosstab(df[feature], df['checkout_status'])
    chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
    chi2_stats.append(chi2_stat)
    p_values_chi2.append(p_value)

# Create a DataFrame to display the results
chi2_test_results = pd.DataFrame({
    'Feature': cat_features,
    'Chi2-statistic': chi2_stats,
    'p-value': p_values_chi2
})

t_test_results, chi2_test_results
```
